```python
import os
import numpy as np
```

```python
E_array = []
with open('generation_data/E.dat', 'r') as f:
    for line in f:
        E_array.append([float(x) for x in line.split()])

E_array = np.array(E_array)
E_array.shape
```

```python
np.unique(E_array)
```

```python
E_array = np.array([np.unique(line) for line in E_array]).reshape(-1)
```

```python
layers = []
for i in range(1, len(E_array)):
    if E_array[i] != E_array[i-1]:
        layers.append([10 * i, E_array[i-1]])
layers.append([10 * i, E_array[-1]])
```
